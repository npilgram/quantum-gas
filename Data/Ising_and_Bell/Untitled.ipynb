```python
import numpy as np
import scipy 
import scipy.signal
import random
import numpy.fft
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
%matplotlib inline
exp = np.exp
arange = np.arange
```

```python
def sym_derivative(x,y,smooth=None):
    #if smooth=='yes':
     #   y_in=scipy.signal.savgol_filter(y,51,3)
    #elif smooth=='no':
      #  y_in=y
    deriv=np.zeros(len(x))
    for i in range(len(x)):
        if i==0:
            dy=y[i+1]-y[i]
            dx=x[i+1]-x[i]
            deriv[i]=dy/dx
        elif i==len(x)-1:
            dy=y[i]-y[i-1]
            dx=x[i]-x[i-1]
            deriv[i]=dy/dx
        else:
            dy=y[i+1]-y[i-1]
            dx=x[i+1]-x[i-1]
            deriv[i]=dy/dx
    return deriv
```

```python
array20=np.loadtxt('Ising_and_Bell_State_4.3_1.3_920_50000_tau1_4.6_40x40_1.txt')
T20=array20[0]
E20=array20[1]
CV20=array20[2]
M20=array20[3]
Chi20=array20[4]
ED20=array20[5]
D20=array20[6]
E_smooth20=scipy.signal.savgol_filter(E20,51,3)
CV20_deriv=sym_derivative(T20,E_smooth20)

array40=np.loadtxt('Ising_and_Bell_State_4.3_1.3_920_50000_tau2_460_40x40_1.txt')
T40=array40[0]
E40=array40[1]
CV40=array40[2]
M40=array40[3]
Chi40=array40[4]
ED40=array40[5]
D40=array40[6]
E_smooth40=scipy.signal.savgol_filter(E40,51,3)
CV40_deriv=sym_derivative(T40,E_smooth40)

plt.plot(T20,M20,'ro')
plt.plot(T40,M40,'bo')
plt.show()
```
